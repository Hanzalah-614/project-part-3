**Intro:**
This article talks about the how competition for food and nest sites leads to aggressive behavior in bird species, with the model organism being Collared Flycatcher (Ficedula albicollis). Researchers used three mock (not living animals, stuffed specimens) competitor species: Great tits (food and nest competitor), Chaffinch (food competitor) and Dunnock (neither food nor nest competitor, control species) to simulate the presence of real competitors and analyze the varying levels of competition and their impact on the aggressiveness of Collard Flycathers. The purpose fo these findings are to understand how resource competition influences behavior and contributes to ecological relationships.

I chose a barplot for "breeding stage" because it shows the frequency of contact (aggression by making contact) of the the different breeding stages

```python
#make a barplot with the df "dataset" and use 'breeding stage' for the x-axis and 
#mean contact for y-axis
sns.barplot(data = dataset,
            x = 'breeding stage',
            y = 'contact?')
```
